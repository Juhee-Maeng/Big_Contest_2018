```python
import pandas as pd
import numpy as np

# Configuring pandas library
pd.options.display.max_columns = 100
pd.options.display.precision = 4
```

```python
path = 'data/final_data_rev/train/'
DF_label = pd.read_csv(path+'train_label.csv')
DF_activity = pd.read_csv(path+'train_activity.csv')
```

```python
DF_activity.head()
```

```python
week = 1
DF_week_activity = DF_activity[DF_activity['wk'] == week]

DF_week_activity_groupby = DF_week_activity.groupby('acc_id').agg('sum')
DF_week_activity_groupby = DF_week_activity_groupby.reset_index()
DF_week_activity_groupby = DF_week_activity_groupby.drop('wk', axis = 1)

print("the number of player on {} week : {}".format(week, DF_week_activity_groupby.shape[0]))

col = DF_week_activity_groupby.columns.tolist()

for i in range(len(col)):
    if (col[i] != 'acc_id'):
        col[i] = 'wk'+str(week)+'_'+col[i]  

DF_week_activity_groupby.columns = col

DF_merge = pd.merge(left = DF_label, right = DF_week_activity_groupby, how='left', on = 'acc_id')
```
