```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import minmax_scale
pd.options.display.max_columns = 403
```

```python
path = 'data/final_data_rev/train/'
DATA = pd.read_csv(path + 'train_ver1_1.csv')
col_list = DATA.columns

basis = ['acc_id', 'label', 'guild', 'wk1_cnt_dt', 'wk2_cnt_dt', 'wk3_cnt_dt', 
         'wk4_cnt_dt', 'wk5_cnt_dt', 'wk6_cnt_dt', 'wk7_cnt_dt', 'wk8_cnt_dt']
DF = DATA.drop(labels=basis, axis = 1)
DF = DF.replace({0:-1})
```

```python
num = 100000
for col_name in DF.columns.tolist():
    empty = np.zeros(num) - 1
    target = DF[col_name]
    idx = target[target.values != -1].index
    empty[idx] = minmax_scale(target[idx])
    DF[col_name] = empty
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
X = DATA.drop(['acc_id', 'label'],axis = 1).values
y = DATA['label'].values

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, random_state = 1)

model1 = RandomForestClassifier(n_estimators = 100)
model1.fit(X_train, y_train)
print("TRAIN : ", f1_score(y_train, model1.predict(X_train), average = 'macro'))
print("VALID : ", f1_score(y_val, model1.predict(X_val), average = 'macro'))
```

```python
DATA2=pd.concat([DATA[basis], DF], axis = 1)[col_list]

X = DATA2.drop(['acc_id', 'label'], axis = 1).values
y = DATA2['label'].values

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, random_state = 1)

model2 = RandomForestClassifier(n_estimators = 100)
model2.fit(X_train, y_train)
print("TRAIN : ", f1_score(y_train, model2.predict(X_train), average = 'macro'))
print("VALID : ", f1_score(y_val, model2.predict(X_val), average = 'macro'))
```
