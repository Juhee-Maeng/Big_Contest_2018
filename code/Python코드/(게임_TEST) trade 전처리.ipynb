```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'data/final_data_rev/test/'
DF = pd.read_csv(path + 'test_trade.csv')
DF_L=pd.read_csv(path + 'test_label.csv')
```

```python
IT=['grocery','money','weapon','gem','costume','accessory']
```

```python
DF=DF.drop(['trade_day','trade_time'],axis=1)
```

```python
for week in range(1,9):
    DF_W=DF[DF['trade_week']==week]
    for type in IT:
        DF_W_it=DF_W[DF_W['item_type']==type]
        DF_W_it_S=DF_W_it.groupby('source_acc_id').agg('sum')
        DF_W_it_S=DF_W_it_S.drop('trade_week',axis=1)
        DF_W_it_S=DF_W_it_S.reset_index()
        DF_W_it_S.columns=['acc_id', 'wk' + str(week)+'_source_'+type]
        DF_L = pd.merge(left = DF_L, right = DF_W_it_S, on='acc_id', how='left')
        DF_W_it_T=DF_W_it.groupby('target_acc_id').agg('sum')
        DF_W_it_T=DF_W_it_T.drop('trade_week',axis=1)
        DF_W_it_T=DF_W_it_T.reset_index()
        DF_W_it_T.columns=['acc_id','wk' + str(week)+'_target_'+type]
        DF_L = pd.merge(left = DF_L, right = DF_W_it_T, on='acc_id', how='left')
```
