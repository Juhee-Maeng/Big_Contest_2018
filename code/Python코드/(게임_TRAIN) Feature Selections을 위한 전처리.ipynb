```python
import pandas as pd
import numpy as np
import datetime

pd.options.display.max_columns = 53
path1 = 'data/final_data_rev/train/'
path2 = 'data/final_data_rev/test/'
DATA = pd.read_csv(path1+'train_ver1_3.csv')
TEST = pd.read_csv(path2+'test_ver1_3.csv')
display(DATA.head())
print('data shape : ', DATA.shape)
```

```python
lst_max = []
lst_min = []
lst_mean = []
now = datetime.datetime.now()
for i in range(40000):
    acc_id = TEST.loc[i,'acc_id']
    data = TEST[TEST['acc_id'] == acc_id]
    lst_max.append(data.max().values)
    lst_min.append(data.min().values)
    lst_mean.append(data.mean().values)
print(now-datetime.datetime.now())
```

```python
col_list = DATA.columns.tolist()
max_col = []
min_col = []
mean_col = []
for i in range(0,53):
    item = col_list[i]
    if (item == 'acc_id') :
        max_col.append(item)
        min_col.append(item)
    elif (item == 'label') :
        max_col.append(item)
        min_col.append(item)
        mean_col.append(item)
    elif (item == 'guild') :
        max_col.append(item)
        min_col.append(item)
        mean_col.append(item)
    else:    
        max_col.append('max_'+item)
        min_col.append('mix_'+item)
        mean_col.append('mean_'+item)
        
DF_max = pd.DataFrame(lst_max)
DF_max.columns = max_col

DF_min = pd.DataFrame(lst_min)
DF_min.columns = min_col

DF_mean = pd.DataFrame(lst_mean)
DF_mean.columns = mean_col
DF_mean['acc_id'] = TEST['acc_id']

DF_FULL = pd.concat([DF_max, DF_min.drop(['acc_id','label','guild'],axis=1)], axis = 1)
DF_FULL = pd.concat([DF_FULL, DF_mean.drop(['acc_id','label','guild'],axis=1)], axis = 1)
DF_FULL.to_csv(path2+'test_feature.csv', index=False,encoding=False)
```
