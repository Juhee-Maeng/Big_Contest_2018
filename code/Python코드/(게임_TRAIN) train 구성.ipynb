```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

# Configuring pandas library
pd.options.display.max_columns = 50
pd.options.display.precision = 3
```

```python
path = 'data/final_data_rev/train/'
ACTIVITY = pd.read_csv(path+'activity.csv')
PARTY = pd.read_csv(path+'party.csv')
GUILD = pd.read_csv(path+'guild.csv')
TRADE = pd.read_csv(path+'trade.csv')
PAYMENT = pd.read_csv(path+'payment.csv')
LABEL = pd.read_csv(path+'train_label.csv')
```

```python
ACTIVITY = ACTIVITY.drop('label', axis = 1)
ACTIVITY = ACTIVITY.fillna(0)
print(ACTIVITY.shape)
ACTIVITY.head()
```

```python
PARTY = PARTY.fillna(0)
PARTY = PARTY.drop('label', axis = 1)
PARTY.columns = ['acc_id', 'wk1_party', 'wk2_party', 'wk3_party','wk4_party', 
                 'wk5_party', 'wk6_party', 'wk7_party', 'wk8_party',]
display(PARTY.head())
PARTY.shape
```

```python
GUILD = GUILD.drop('label', axis = 1)
display(GUILD.head())
GUILD.shape
```

```python
TRADE = TRADE.drop('label', axis = 1)
TRADE = TRADE.fillna(0)
display(TRADE.head())
TRADE.shape
```

```python
PAYMENT = PAYMENT.drop(['Unnamed: 0', 'label'], axis = 1)
PAYMENT.columns = ['acc_id', 'wk1_payment', 'wk2_payment', 'wk3_payment', 'wk4_payment', 
                   'wk5_payment', 'wk6_payment', 'wk7_payment', 'wk8_payment']

minimum = PAYMENT.iloc[:,1:].min().min()
PAYMENT = PAYMENT.replace({minimum : 0})

display(PAYMENT.head())
PAYMENT.shape
```

```python
'''
DF_train = ACTIVITY.copy()
DF_train = pd.merge(left = DF_train, right = PARTY, on = 'acc_id', how = 'left')
DF_train = pd.merge(left = DF_train, right = GUILD, on = 'acc_id', how = 'left')
DF_train = pd.merge(left = DF_train, right = TRADE, on = 'acc_id', how = 'left')
DF_train = pd.merge(left = DF_train, right = PAYMENT, on = 'acc_id', how = 'left')
display(DF_train.head())
DF_train.shape
'''
DF_train = LABEL.copy()
DF_train = pd.merge(left = DF_train, right = GUILD, on = 'acc_id', how = 'left')
DF_train = pd.merge(left = DF_train, right = ACTIVITY, on = 'acc_id', how = 'left')
DF_train = pd.merge(left = DF_train, right = PARTY, on = 'acc_id', how = 'left')
DF_train = pd.merge(left = DF_train, right = TRADE, on = 'acc_id', how = 'left')
DF_train = pd.merge(left = DF_train, right = PAYMENT, on = 'acc_id', how = 'left')
display(DF_train.head())
DF_train.shape
```

```python
# 0820 update : code for scaling
from sklearn.preprocessing import robust_scale
from sklearn.preprocessing import minmax_scale
num = 100000
for col in DF_train.columns:
    if (col != 'acc_id') and (col != 'label'):
        zeros = np.zeros(num) # make zero array

        temp = DF_train[col]
        idx = temp[temp.values != 0].index

        zeros[idx] = minmax_scale(temp.values[idx].reshape(-1,1)).reshape(1,-1)
        
        DF_train[col] = zeros
```

```python
DF_train.to_csv(path+'train_ver1_2.csv', index = False, encoding = False)
```
