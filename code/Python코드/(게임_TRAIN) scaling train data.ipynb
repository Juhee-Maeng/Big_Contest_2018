```python
# Importing Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
pd.options.display.max_columns = 53

path = 'data/final_data_rev/test/'
DF = pd.read_csv(path+'test_ver1_1.csv')
length = DF.shape[0]

col_list = DF.columns.tolist()
weekend = ['wk1', 'wk2', 'wk3', 'wk4', 'wk5', 'wk6', 'wk7', 'wk8']
basis = ['acc_id', 'label', 'guild']
dic = {'wk1' : [], 'wk2' : [], 'wk3' : [], 'wk4' : [], 'wk5' : [], 'wk6' : [], 'wk7' : [], 'wk8' : []}

for wk in weekend:
    for col in col_list:
        if wk in col:
            dic[wk].append(col)

DF_wk1 = pd.concat([DF[basis], DF[dic['wk1']]], axis = 1)
DF_wk2 = pd.concat([DF[basis], DF[dic['wk2']]], axis = 1)
DF_wk3 = pd.concat([DF[basis], DF[dic['wk3']]], axis = 1)
DF_wk4 = pd.concat([DF[basis], DF[dic['wk4']]], axis = 1)
DF_wk5 = pd.concat([DF[basis], DF[dic['wk5']]], axis = 1)
DF_wk6 = pd.concat([DF[basis], DF[dic['wk6']]], axis = 1)
DF_wk7 = pd.concat([DF[basis], DF[dic['wk7']]], axis = 1)
DF_wk8 = pd.concat([DF[basis], DF[dic['wk8']]], axis = 1)

column = ['acc_id', 'label', 'guild', 'cnt_dt', 'play_time', 'npc_exp', 'npc_hongmun', 'quest_exp', 'quest_hongmun', 'item_hongmun', 
          'game_combat_time', 'get_money', 'duel_cnt', 'duel_win', 'partybattle_cnt', 'partybattle_win', 'cnt_enter_inzone_solo', 
          'cnt_enter_inzone_light', 'cnt_enter_inzone_skilled', 'cnt_enter_inzone_normal', 'cnt_enter_raid', 'cnt_enter_raid_light', 
          'cnt_enter_bam', 'cnt_clear_inzone_solo', 'cnt_clear_inzone_light', 'cnt_clear_inzone_skilled', 'cnt_clear_inzone_normal', 
          'cnt_clear_raid', 'cnt_clear_raid_light', 'cnt_clear_bam', 'normal_chat', 'whisper_chat', 'district_chat','party_chat', 
          'guild_chat', 'faction_chat', 'cnt_use_buffitem', 'gathering_cnt', 'making_cnt','party', 'source_grocery', 'target_grocery',
          'source_money', 'target_money', 'source_weapon', 'target_weapon', 'source_gem', 'target_gem', 'source_costume', 'target_costume', 
          'source_accessory', 'target_accessory', 'payment']

DF_wk1.columns = column
DF_wk2.columns = column
DF_wk3.columns = column
DF_wk4.columns = column
DF_wk5.columns = column
DF_wk6.columns = column
DF_wk7.columns = column
DF_wk8.columns = column
```

```python
DF_full = pd.concat([DF_wk1, DF_wk2, DF_wk3, DF_wk4, DF_wk5, DF_wk6, DF_wk7, DF_wk8], ignore_index=True)
```
